Load the pickle file data

```python
import pickle

with open('/Users/henryfriesen/LHL_projects/capstone_storage_files/training_results.pkl','rb') as f:
    training_results = pickle.load(f)

epoch_loss_train = training_results['epoch_loss_train']
epoch_loss_test = training_results['epoch_loss_test']
epoch_acc_train = training_results['epoch_acc_train']
epoch_acc_test = training_results['epoch_acc_test']
epoch_learning_rate = training_results['epoch_learning_rate']
```

Plot results

```python
import matplotlib.pyplot as plt
import numpy as np

N_epoch = len(epoch_loss_train)

fig, ax = plt.subplots(figsize=(15,5),ncols=3)

ax[0].plot(np.arange(0,N_epoch),epoch_loss_train,np.arange(0,N_epoch),epoch_loss_test)
ax[0].set_title('Loss')
ax[1].plot(np.arange(0,N_epoch),epoch_acc_train,np.arange(0,N_epoch),epoch_acc_test)
ax[1].set_title('Accuracy')
ax[2].plot(np.arange(0,N_epoch),epoch_learning_rate)
ax[2].set_title('Learning Rate')

plt.show()
```
