```python
total_all_count = 0
total_hem_count = 0
for path, directories, files in os.walk('/Users/henryfriesen/LHL_projects/Capstone/C-NMC_Leukemia'):
     for file in files:
        all_count = 0
        hem_count = 0
        if "training" in path:
            if "all" in path:
                all_count = len(files)
            elif "hem" in path:
                hem_count = len(files)
            total_all_count += all_count
            total_hem_count += hem_count
            break
print(f"Cell count of normal cells: {total_hem_count} \nCell count of Leukemic cells: {total_all_count}")
```

```python
# Seaborn plots
# need to edit into seaborn version. 

data = {'HEM - Normal Cells': total_hem_count, 'ALL - Leukemic Cells': total_all_count}
keys = data.keys()
values = data.values()

# fig = sns.barplot(data=data, x=keys, y=values)
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(keys, values, color ='navy')
 
plt.xlabel("Class")
plt.ylabel("Count")
plt.title("Training Data Class Numbers")
plt.show()
```
